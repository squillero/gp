```python
import operator
```

```python
from gxgp import Node

tree = Node(operator.mul, [Node(operator.add, [Node(10),
                                               Node('x')]),
                           Node(2)])
tree.draw()
```

```python
tree(x=11)
```

```python
tree()  # Error: missing 'x'
```

```python
tree2 = Node(lambda a, b, c, d, e: (a + b * c - d) / e,
             [Node('x'), Node('y'), Node(3), Node('z'), Node(2)])
tree2.draw()
```

```python
# Explicit keyword arguments (unused arguments are ignored)
tree2(x=1, y=2, z=3, foo=42)
```

```python
# Unpacking a dictionary
vars = {'x': 1, 'y': 2, 'z': 3, 'foo': 42}
tree2(**vars)
```
